```python
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def segment(Image):
    img = plt.imread(Image)
    m = img.shape[0]
    n = img.shape[1]
        
    for i in range(m):
        for j in range(n):
            for k in range(3):
                if img[i][j][k] < 0.5:
                    img[i][j][k] = 0
                else:
                        img[i][j][k] = 1
            
    for i in range(m):
        for j in range(n):
            if img[i][j][0] != 1 or img[i][j][1] != 1 or img[i][j][2] != 1:
                img[i][j][0] = 0
                img[i][j][1] = 0
                img[i][j][2] = 0
                
    try:
        if not os.path.exists('Output'):
            os.makedirs('Output')

        plt.imsave("./Output/output" + Image[11:], img)
            
    except OSError: 
        print ('Error: Creating directory of data')

    plt.figure(figsize=(10,8))
    plt.imshow(img)
    plt.show()
```

```python
for i in range(1, 6):
    segment('Input/input' + str(i) + '.png')
```
